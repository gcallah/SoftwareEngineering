Now let's get the total for all '5' cols: this will be the best 3 scores out of the 4 cols. First we just add the 4:

```python
TOTAL_5 = '5 total'
df[TOTAL_5] = df['5.1'] + df['5.2'] + df['5.3'] + df['5.4']
```

```python
df.head(10)
```

Now create a column that is the min of the 5s:

```python
min5 = df[['5.1', '5.2', '5.3', '5.4']].min(axis=1)
idxmin5 = df[['5.1', '5.2', '5.3', '5.4']].idxmin(axis=1)
df['min5'] = min5
df['idxmin5'] = idxmin5
```

```python
df.head(10)
```

Now subtract min5 from TOTAL_5:

```python
df[TOTAL_5] -= df['min5']
```

```python
df.head(10)
```

Now caclulate the total score:

```python
df['Total'] = (df['3.1']
               + df['3.2']
               + df['3.3']
               + df['3.4']
               + df['3.5']
               + df['3.6']
               + df['3.7']
               + df['3.8']
               + df['3.9']
               + df['3.10']
               + df['3.11']
               + df['3.12']
               + df['4.1']
               + df['4.2']
               + df['4.3']
               + df['4.4']
               + df['4.5']
               + df['4.6']
               + df['4.7']
               + df['4.8']
               + df[TOTAL_5]
              )
```
